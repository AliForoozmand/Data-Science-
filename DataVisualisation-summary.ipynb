## Plotting real valued functions

Any real valued function $f$ defined of the interval $(a,\,b)$ is, in principle defined by an infinite amount of values, namely $f(x)$ for every $x\in(a,\,b)$. All these values can not be represented numerically, given that every computer has finite memory. As such, in data science, we choose a sampling of points $x_i$ for $i=0,\,...\,,N-1$, and represent the function by evaluating the function on these sampling points, $f_i = f(x_i)$. Both the sampling points as well as the function evaluation are saved as numpy array, which facilitate quick manipulations without having to loop through all entries. A very common way of sampling an interval $(a,\,b)$ with $N$ points, is to space them at equal distances with $\texttt{numpy.linspace(a, b, num=N)}$.

## Plotting measured data and model prediction

$\textbf{Case: }$ In scientific amplications, measurements often come with associated errorbars, that we need to represent in our plots. Similarly, theoretical predictions might be uncertain. Lets consider the simplified case of calibration a spring balance (German federwaage). For a set of standard weights $w_i$, we record elongations of the spring $l_i$, with $i=0,\,...,\, N-1$, for which we estimate an error $\delta l_i$. From this data, we can calibrate the relation between weight $w$ and elongation $l$ as $l=\alpha w$, which $\alpha=1 \pm 0.1$ mm/kg.

```python
# data generation
alpha = 1
w = np.array([10, 20, 30, 50, 75, 100, 200, 500])
dl = 2
rel_err = 0.1

errtot = (dl+rel_err*w)

l = alpha*w+np.random.randn(w.shape[0])*errtot

data = np.vstack((w, l, errtot)).T

np.savetxt('spring_balance_calibration.txt', data,
           header='weight[kg],elongation[mm],error[mm]', delimiter=",")
```

$\textbf{Task:}$ Load the data $\texttt{spring_balance_calibration.txt}$ and plot it using $\texttt{\plt.errorbar}$ ($\textit{Hint:}$ find the documentation for loading $\texttt{.txt}$ files and $\texttt{\plt.errorbar}$  online). Do not forget to label both axis with the respective units! Does the plot look better spaced with logarithmic scale on both axis?

```python
w, l, errtot = np.genfromtxt('spring_balance_calibration.txt',
                            delimiter=',').T

plt.figure()
plt.errorbar(w, l, yerr=errtot, fmt='*')
plt.xscale('log')
plt.yscale('log')
plt.xlabel('weight [kg]')
plt.ylabel('elongation [mm]')
```

$\textbf{Task:}$ Use $\texttt{plt.fill$\_$between}$ to shade the band encompassed by $l=(\alpha-\delta\alpha) w$ and $l=(\alpha+\delta\alpha) w$ to show the theoretical uncertainty of the calibration. For aestetical reasons, create some transparency by use the argument $\texttt{alpha=0.2}$)

```python
w_arr = np.geomspace(w[0], w[-1], num=100)

alpha = 1
dalpha= 0.1

l_mean = alpha*w_arr
l_upper = (alpha+dalpha)*w_arr
l_down = (alpha-dalpha)*w_arr

plt.figure()
plt.errorbar(w, l, yerr=errtot, fmt='*', label='data')
plt.xscale('log')
plt.yscale('log')
plt.xlabel('weight [kg]')
plt.ylabel('elongation [mm]')

plt.plot(w_arr, l_mean, color='C1')
plt.fill_between(w_arr, l_down, l_upper, color='C1',
                alpha=0.5, label='model')
plt.legend(loc='upper left')
```

$\textbf{Task:}$ Load the file $\texttt{mass_mass.npy}$ and display it. Add a colorbar to it. $\textit{Hint:}$ Considerig the large dynamical range of the image, use the option $\texttt{norm=LogNorm()}$ let the color scale off the logrithm of the image value. $\textit{Background:}$ This map is part of a scientific study I lead (https://arxiv.org/pdf/2103.16212.pdf), in which we analysed the typical mass distribution of the most massive structures in the Universe, galaxy clusters. This is drawn from a simulation of a large cosmic volume (https://www.tng-project.org/).

```python
# Define the size of the mass map (grid size)
grid_size = 100

# Generate a synthetic mass distribution (random clusters with Gaussian profiles)
def generate_mass_map(grid_size, num_clusters):
    mass_map = np.zeros((grid_size, grid_size))
    
    for _ in range(num_clusters):
        # Randomly choose cluster center positions
        x_center = np.random.randint(0, grid_size)
        y_center = np.random.randint(0, grid_size)
        
        # Randomly assign mass to the cluster
        cluster_mass = np.random.uniform(1e13, 1e15)  # in solar masses
        
        # Create a Gaussian profile for the mass distribution around the center
        for i in range(grid_size):
            for j in range(grid_size):
                dist = np.sqrt((i - x_center) ** 2 + (j - y_center) ** 2)
                mass_map[i, j] += cluster_mass * np.exp(-dist**2 / (2 * (grid_size / 10) ** 2))  # Gaussian spread
    
    return mass_map

# Parameters
num_clusters = 10  # Number of galaxy clusters

# Generate the mass map
mass_map = generate_mass_map(grid_size, num_clusters)

# Save the mass map as a .npy file
np.save('mass_map.npy', mass_map)

# Optional: Visualize the mass map
plt.imshow(mass_map, cmap='inferno', origin='lower')
plt.colorbar(label='Mass (solar masses)')
plt.title('Synthetic Mass Distribution of Galaxy Clusters')
plt.show()
```

It can be very helpful to visualize your 2d arrays also with contour plots. Contour plots show countour $\mathcal{C}_l$ lines associated to a level $l$, which for a 2d function $f(x,y)$ is defined as the set of point $(x,\,y)$ for which $f(x,\,y)=l$. Typical, several levels are chosen to highlight the salient features of your 2d function. A typical examples are contour lines in cartography or isobars in meterology.

$\textbf{Homework:}$ Visualize the Rosenbrock function $f(x,\,y) = (a-x)^2 + b\,(y-x^2)^2$ for $a=1$ and $b=100$. Inform yourself why this function is interesting. $\textit{Hint:}$ the function $\texttt{np.meshgrid}$ will be very helpful in this endevour.

```python
def rosenbrock(x, y, a=1, b=100):
    return (a - x)**2 + b * (y - x**2)**2

x = np.linspace(-2, 2, 400)
y = np.linspace(-1, 3, 400)
X, Y = np.meshgrid(x, y)
Z = rosenbrock(X, Y)


plt.figure(figsize=(10, 8))
contour = plt.contourf(X, Y, Z, levels=50, cmap='viridis')
plt.colorbar(contour)
plt.title('Rosenbrock Function')
plt.xlabel('x')
plt.ylabel('y')
plt.plot(1, 1, 'r*', markersize=10)
plt.annotate('   Global minimum (1,1)', xy=(1, 1))
plt.show()
```
